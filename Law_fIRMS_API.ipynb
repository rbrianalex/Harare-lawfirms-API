```python
#  endpoint for the Places API nearby search
endpoint = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

#  parameters for the API request
params = {
    "location": "-17.8252,31.0335",
    "radius": "10000",
    "keyword": "law firms",
    "key": API_KEY
}

# Make the API request
response = requests.get(endpoint, params=params)
data = response.json()

# Create a GeoJSON FeatureCollection
feature_collection = {
    "type": "FeatureCollection",
    "features": []
}
# Iterate through the places and add them as features
for place in data["results"]:
    feature = {
        "type": "Feature",
        "geometry": {
            "type": "Point",
            "coordinates": [place["geometry"]["location"]["lng"], place["geometry"]["location"]["lat"]]
        },
        "properties": {
            "name": place["name"],
            "address": place.get("vicinity", ""),
            # Add more properties as needed
        }
    }
    feature_collection["features"].append(feature)

# Convert the FeatureCollection to GeoJSON string
geojson_str = json.dumps(feature_collection, indent=2)

# Write the GeoJSON string to a file
with open("law_firms.geojson", "w") as f:
    f.write(geojson_str)

print("GeoJSON file has been created successfully.")
```

```python
# Load the GeoJSON file
gdf = gpd.read_file("/content/law_firms.geojson")

# Create a folium map centered around Harare, Zimbabwe
m = folium.Map(location=[-17.8252, 31.0335], zoom_start=13)

# Add the GeoJSON data to the map with red stars as markers
for idx, row in gdf.iterrows():
    folium.Marker(
        location=[row.geometry.y, row.geometry.x],
        icon=folium.Icon(color='red', icon='star')
    ).add_to(m)

# Display the map
m
```
